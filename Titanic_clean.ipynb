```python
#Clean the data
#Change all column names to small letters
titanic.columns
titanic.columns = titanic.columns.str.lower()
titanic.columns
```

```python
titanic
```

```python
import numpy as np
#created new column
titanic["Fate"] = np.where(titanic["survived"] == 0, "Died", "Survived")
titanic.Fate
```

```python
titanic["pclass"] = titanic["pclass"].replace(3, "Third Class")
titanic["pclass"] = titanic["pclass"].replace(2, "Second Class")
titanic["pclass"] = titanic["pclass"].replace(1, "First Class")

titanic
```

```python

#fill missing value in age column to average age 29.699118
titanic.age = titanic.age.fillna(29.699118)
titanic
titanic.info()
```

```python
#convert age column to integer
titanic.age = titanic.age.astype(int)
titanic
titanic.info()
```

```python
#split the name

titanic[["first_name", "last_name"]] = titanic.name.str.split(",", expand = True) #the double [] was used because we are changing/creating more than one column
titanic
```

```python
titanic.last_name.str.split(".", n=1, expand = True) #the n=1 expands it just once just to have two columns
titanic[["name_prefix", "other_names"]] = titanic.last_name.str.split(".", n=1, expand = True)
titanic
```

```python
titanic = titanic.drop('survived', axis=1)
titanic

```

```python
titanic = titanic.drop(["ticket", "last_name"], axis=1)
titanic
```
