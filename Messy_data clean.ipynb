```python
import pandas as pd

messy = pd.read_csv("C:/Users/deaja/OneDrive/Desktop/toludrive/messy_data.csv", sep='|', encoding='latin-1')
messy
```

```python
messy2 = messy.iloc[:, 0].str.split(';', expand=True) #iloc selects certain rows and columns


```

```python
messy2.drop(8, inplace=True, axis=1)
```

```python
messy2.columns = ['IMBD title ID', 'Original title', 'Release year', 'Genre', 'Duration', 'Country', 'Content Rating', 'Director', 'Income', 'Votes', 'Score']

```

```python
messy2.columns = messy2.columns.str.lower()
messy2
```

```python
messy2.info()
```

```python
messy2['release year'] = pd.to_datetime(messy2['release year'],errors='coerce',infer_datetime_format=True).dt.strftime('%d-%m-%Y')
```

```python
messy2.info()
```

```python
messy2.rename(columns={'release year': 'release date'}, inplace=True)
```

```python
messy2
```

```python
import re
messy2['score'] = messy2['score'].apply(lambda x: re.sub(r'[^\d.]+', '', str(x)))

```

```python
import numpy as np
def convert_to_float(value):
    try:
        return float(value)
    except ValueError:
        return np.nan
messy2['score'] = messy2['score'].apply(convert_to_float)
```

```python
messy2.to_csv('messy_cleaned.csv', index=False)
```
